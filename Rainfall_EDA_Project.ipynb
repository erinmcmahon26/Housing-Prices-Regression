Due to the differences in water need across India, there is no agreed upon average water used per person. This article, https://www.researchgate.net/publication/334807513_An_Analysis_of_Demand_and_Supply_of_Water_in_India, states that the water need in 2010 was 710 bcm. According to my dataset there was about 5285 bcm (originally 52855 mm). If all rainfall was able to be collected it would have been enough to supply the country with its water needs. According to this article, "only 48% of rainfall ends up in Indiaâ€™s rivers. Due to lack of storage and crumbling infrastructure, only 18% can be utilized. Rainfall is confined to the monsoon season, June through September, when India gets, on average, 80% of its total annual precipitation." With this being applied to the numbers, only about 2537 bcm of rainfall was collected in India's rivers and only 951 bcm can be utilized. Most of this water is collected during the monsoon season so based off this data and article, the country is being supplied with enough water for it's population but most of it can only be collected during one part of the year, Jun-Sept.
